# Création du fichier d'export des résultats

Le fichier ne contient que les DPE a mettre à jour (les résultats identiques au rappro de 2024 sont exclus)

```python
import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
# Export pour MAJ de la base DPE
# On n'exporte que les lignes
#   - avec un seul ID-RNB trouvé par le rapprochement
#   - qui n'ont pas d'id-rnb deja rempli avec une provenance logicielle
#   - qui n'ont pas un id-rnb avec provenance reprise RNB + identifiant similaire

df_export = pd.read_csv('notebooks/rapprochements/DPE/2025/results_DPE_RNB.csv')
```

```python
# la colonne rnb_id_rappro est considérée comme une colonne de texte. Un id-rnb unique est stocké sous la forme ['XXXAAAXXXAAA'] soit 16 caractères
df_export_filtered = df_export[(df_export['rnb_id_rappro'].apply(len) == 16) & (df_export['provenance_id_rnb'] != 'Logiciel')]

df_export_filtered['rnb_id_rappro'] = df_export_filtered['rnb_id_rappro'].apply(lambda id: id.strip("'[]"))
df_export_filtered = df_export_filtered[df_export_filtered['rnb_id_rappro'] != df_export_filtered['id_rnb']]
```

```python
# par curiosité, on regarde combien de rnb_id sont assignés à des DPE qui n'en avaient pas 
# et combien sont des mises à jour de rnb_id (ce qui signifie probablement que le RNB a été modifié et est maintenant plus précis que l'année dernière)
df_export_filtered['rnb_update'] = df_export_filtered['id_rnb'].notna()
df_export_filtered.groupby('rnb_update').count()
```

```python
# export du fichier
df_export_filtered[['numero_dpe', 'file', 'rnb_id_rappro']].to_csv('notebooks/rapprochements/DPE/2025/update_rapprochement_DPE_RNB.csv', index=False)
```
