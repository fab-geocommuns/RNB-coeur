```python
from batid.services.geocoders import BanBatchGeocoder
from io import StringIO

df_final = pd.read_csv('./data/geocode_this.csv')
temp_filename = './data/geocode_tmp.csv'
batch_size = 500
all_batches = []

for i in range(0, len(df_final), batch_size):
    batch_df = df_final.iloc[i:i + batch_size]
    batch_df.to_csv(temp_filename, index=False)
    
    geocoder = BanBatchGeocoder()
    response = geocoder.geocode_file(temp_filename, columns=['adresse'], citycode_col='depcom')
    
    response_csv = StringIO(response.text)
    df_geocoded_batch = pd.read_csv(response_csv, sep=',')
    
    all_batches.append(df_geocoded_batch)

df_geocoded = pd.concat(all_batches, ignore_index=True)
df_geocoded.to_csv('./data/geocoded.csv', index=False)
```

```python
from django.db import connection

def get_rnb_id_from_point(row, buffer=0):
    rawSql = f"""
        select *
        from batid_building
        where st_intersects(shape, ST_Buffer(ST_GeomFromText('POINT({row['lon']} {row['lat']})', 4326)::geography, {buffer}))
    """

    with connection.cursor() as cursor:
        cursor.execute(rawSql)
        results = cursor.fetchall()
        return [b[1] for b in results]
```

```python
def get_rnb_id_from_point_ril(row, buffer=0):
    rawSql = f"""
        select *
        from batid_building
        where st_intersects(shape, ST_Buffer(ST_GeomFromText('POINT({row['longitude']} {row['latitude']})', 4326)::geography, {buffer}))
    """

    with connection.cursor() as cursor:
        cursor.execute(rawSql)
        results = cursor.fetchall()
        return [b[1] for b in results]
```

```python
from batid.services.guess_bdg_new import ClosestFromPointHandler
from batid.services.guess_bdg_new import Guesser

guesser = Guesser()
# we only set only one handler
guesser.handlers = [ClosestFromPointHandler()]
```

```python
df_prepare_rapprochement.loc[:, ['ext_id']] = df_prepare_rapprochement['id_ea']
df_for_guesser = df_prepare_rapprochement.copy()
df_for_guesser = df_for_guesser.drop(columns=['lat', 'lon']).rename(columns={'latitude': 'lat', 'longitude': 'lng'})
list_prepare_rapprochement = df_for_guesser.to_dict(orient='records')
```

```python
guesser.load_inputs(list_prepare_rapprochement)
guesser.guess_all()
```

```python
guesser.report()
```

```python
cles_filtrees = [key for key, value in guesser.guesses.items() if len(value['matches']) > 0]
cles_filtrees[0]
```

```python
for k, v in guesser.guesses.items():
    v['rnb_id_from_point'] = [b.rnb_id for b in v['matches']]
```

```python
for k, v in guesser.guesses.items():
    v['ban_id_from_point'] = [address.id for a in v['matches'] for address in a.addresses_read_only.all()]
    
guesser.guesses['751190000002776']
```

```python
all_buildings = []
for k, v in guesser.guesses.items():
    all_buildings.append({
        **v['input'],
        'rnb_id_from_point': v['rnb_id_from_point'],
        'ban_id_from_point': v['ban_id_from_point'],
        'match_reason': v['match_reason'],
    })
```

```python
df_with_building = pd.DataFrame(all_buildings)
df_with_building.head()
```

```python
filtre_1 = df_with_building[df_with_building.apply(lambda x: x['result_id'] not in x['ban_id_from_point'] and len(x['rnb_id_from_point']) > 0, axis=1)]

print("RNB ID détectés via ClosestFromPointHandler qui n'ont pas le BAN ID détecté via geocoding BAN:", len(filtre_1))
```

```python
filtre_2 = df_with_building[df_with_building.apply(lambda x: x['result_id'] in x['ban_id_from_point'], axis=1)]

print("RNB ID détectés via ClosestFromPointHandler qui ont le BAN ID détecté via geocoding BAN:", len(filtre_2))
```
