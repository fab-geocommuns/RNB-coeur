# Rapprochement Data ES

Le fichier data-es.csv contient un miller de lignes d'équipements sportifs, contenant chacun un identifiant et diverses informations, dont des coordonnées (point WGS 84).

Ce petit script trouve le batiment le plus proche dans la base du RNB, et sort l'identifiant data-es, l'identifiant rnb et la distance entre le bâtiment identifié et les coordonnées données initialement.

```python
import csv
from django.db import connection
import os
os.environ["DJANGO_ALLOW_ASYNC_UNSAFE"] = "true"
from batid.models import Building
from django.contrib.gis.geos import GEOSGeometry
from django.contrib.gis.db.models.functions import Distance

with open('data-es.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=';', quotechar='|')
    # skip headers
    next(spamreader)

    for row in spamreader:
        data_es_id = row[0]
        try:
            lon = float(row[6])
            lat = float(row[7])
        except:
            continue


        with connection.cursor() as cursor:
            point = GEOSGeometry(f"POINT ({lon} {lat})", srid=4326)       
            buildings = Building.objects.filter(shape__dwithin=(point, 0.001)).annotate(distance=Distance("shape", point)).order_by("distance")
            print(data_es_id, buildings[0].rnb_id, buildings[0].distance)        
```
