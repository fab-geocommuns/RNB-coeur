## Script de dump des parcelles du village

Script ayant servi au transfert initial des parcelles du village vers un fichier de dump.

````python
### Récupération des ID des parcelles concernées
````

````python
import os
from django.db import connection
from django.core.management import call_command
os.environ["DJANGO_ALLOW_ASYNC_UNSAFE"] = "true"

q = "SELECT id FROM batid_plot WHERE ST_DWithin(shape, st_transform(st_point(5.73592,45.179346, 4326), 2154), 200) OR ST_DWithin(shape, st_transform(st_point(5.6362641,45.0716039, 4326), 2154), 300)"

with connection.cursor() as cur:
    
    cur.execute(q)
    rows = cur.fetchall()
    ids = [r[0] for r in rows]
    
    print(",".join(ids))
    


    
    
````

````python
### Exécution de la commande de dump

Un fois les ids des parcelles concernées récupérées, on peut sortir ces données vers un fichier JSON. 

Pour cela on exécute la commande : 
    
```
python manage.py dumpdata batid.Plot --pks=[toutes les ids séparées par une virgule comme ci-dessus] --output=./batid/fixtures/village_plots.json
```
````
